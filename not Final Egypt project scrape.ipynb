```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time

s = Service('C:/Users/Lenovo/Downloads/chromedriver_win32 (1)/chromedriver.exe')
driver = webdriver.Chrome(service=s)

driver.get("https://egy-map.com/projects?name=&category=0&location=0&case=processing&s=")

SCROLL_PAUSE_TIME = 0.5

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # Scroll down to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python

html = driver.page_source
soup = bs(html, 'lxml')
reslts = soup.findAll('div', {'class': 'single-property-box properties-list-item'})
```

```python
len(reslts)
```

```python
cat = []
name = []
location = []
tim = []
ar = []
cos = []

for x in range(len(reslts)):
    
    name.append(reslts[x].find('div', {'property-title-box'}).h4.text)
    
    locs = reslts[x].find('i', {'fas fa-map-marker-alt'}).next_siblings
    for loc in locs:
        loc = loc.text.strip()
        location.append(loc)
        
        while("" in location) :
            location.remove("")
    
    
    catigories = reslts[x].find('i', {'fas fa-bookmark'}).next_siblings
    for catigory in catigories:
        catigory = catigory.text.strip()
        cat.append(catigory)
        
        while("" in cat) :
            cat.remove("")

            
    try:
        times = reslts[x].find('i', {'class': 'fa fa-clock'}).next_siblings
        for time in times:
            time = time.text.strip()
            tim.append(time)

            while("" in tim):
                tim.remove("")
                
    except:
        tim.append('0')
        
    try:
        areas = reslts[x].find('i', {'class': 'fas fa-arrows-alt'}).next_siblings
        for area in areas:
            area = area.text.strip()
            ar.append(area)
            
            while("" in ar):
                ar.remove("")              
    except:
        ar.append('0')
    
    
    try:
        costs = reslts[x].find('i', {'fas fa-paperclip'}).next_siblings
        for cost in costs:
            cost = cost.text.strip()
            cos.append(cost)
            
            while("" in cos):
                cos.remove("")
    except:
        cos.append('0')
    
df_projects = pd.DataFrame({'Name': name, 'Catigory': cat, 'Location': location, 'ETAs': tim, 'Area': ar, 'Cost': cos})
df_projects
```

```python
df_projects.to_excel('Projects Egypt Scrape.xlsx', index=False)
```
